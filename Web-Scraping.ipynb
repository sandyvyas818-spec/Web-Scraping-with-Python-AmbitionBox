```python
names = []
ratings = []
reviews = []
salaries = []
interviews = []
jobs = []
benefits = []
photos = []
```

```python
for i in company:
    spans = i.find_all('span')

    names.append(i.find('h2').text.strip())
    ratings.append(i.find('div', class_='rating_text').text.strip())

    reviews.append(spans[7].text.strip() if len(spans) > 7 else 'N/A')
    salaries.append(spans[9].text.strip() if len(spans) > 9 else 'N/A')
    interviews.append(spans[11].text.strip() if len(spans) > 11 else 'N/A')
    jobs.append(spans[13].text.strip() if len(spans) > 13 else 'N/A')
    benefits.append(spans[15].text.strip() if len(spans) > 15 else 'N/A')
    photos.append(spans[17].text.strip() if len(spans) > 17 else 'N/A')
```

```python
len(names)
```

```python
df = pd.DataFrame({
    'Company': names,
    'Rating': ratings,
    'Reviews': reviews,
    'Salaries': salaries,
    'Interviews': interviews,
    'Jobs': jobs,
    'Benefits': benefits,
    'Photos': photos})
```

```python
df
```

```python
df.replace(
    ['Reviews','Salaries','Interviews','Jobs','Benefits','Photos',
     'Promotions','Promotions, Salary, Work Satisfaction'],'N/A',
    inplace=True)
```
